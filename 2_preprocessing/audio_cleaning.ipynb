```python
import os
import librosa
import soundfile as sf
import numpy as np
import noisereduce as nr
```

```python
INPUT_DIR = "NJIA/input_audio"
OUTPUT_DIR = "../2_preprocessing/cleaned_audio"

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
TARGET_SR = 16000  # standard for ASR
```

```python
def preprocess_audio(file_path, output_path):
    # Load audio
    audio, sr = librosa.load(file_path, sr=TARGET_SR, mono=True)

    # Noise reduction (light)
    reduced_noise = nr.reduce_noise(
        y=audio,
        sr=TARGET_SR,
        prop_decrease=0.8
    )

    # Normalize
    normalized_audio = librosa.util.normalize(reduced_noise)

    # Save cleaned audio
    sf.write(output_path, normalized_audio, TARGET_SR)
```

```python
os.makedirs(INPUT_DIR, exist_ok=True)

for file_name in os.listdir(INPUT_DIR):
    if file_name.endswith(".mp3") or file_name.endswith(".wav"):
        input_path = os.path.join(INPUT_DIR, file_name)
        output_path = os.path.join(
            OUTPUT_DIR, file_name.replace(".mp3", ".wav")
        )

        preprocess_audio(input_path, output_path)
        print(f"Processed: {file_name}")
```
